```python
import numpy as np
from scipy.spatial.distance import euclidean, cityblock, hamming
from sklearn.metrics import jaccard_score
from numpy import dot
from numpy.linalg import norm

customer_A = np.array([4, 5, 2, 3, 4])
customer_B = np.array([5, 3, 2, 4, 5])

customer_A_binary = np.array([1, 0, 1, 1, 0, 1])
customer_B_binary = np.array([1, 1, 1, 0, 0, 1])


euclidean_dist = euclidean(customer_A, customer_B)

manhattan_dist = cityblock(customer_A, customer_B)
cosine_sim = dot(customer_A, customer_B) / (norm(customer_A) * norm(customer_B))
hamming_dist = hamming(customer_A_binary, customer_B_binary) * len(customer_A_binary)

jaccard_sim = jaccard_score(customer_A_binary, customer_B_binary)

print(" Similarity / Distance Measures:")
print(f"1. Euclidean Distance     : {euclidean_dist:.3f}")
print(f"2. Manhattan Distance     : {manhattan_dist:.3f}")
print(f"3. Cosine Similarity      : {cosine_sim:.3f}")
print(f"4. Hamming Distance       : {hamming_dist:.3f}")
print(f"5. Jaccard Similarity     : {jaccard_sim:.3f}")
```

```python
import numpy as np
from scipy.spatial.distance import chebyshev, minkowski

user1 = np.array([5, 3, 4, 4, 2])
user2 = np.array([4, 2, 5, 4, 3])

chebyshev_dist = chebyshev(user1, user2)

minkowski_dist = minkowski(user1, user2, p=3)

print(" User Rating Similarity Measures:")
print(f"1. Chebyshev Distance   : {chebyshev_dist:.3f}")
print(f"2. Minkowski Distance(p=3): {minkowski_dist:.3f}")
```
