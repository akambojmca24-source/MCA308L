```python
import pandas as pd

df = pd.read_csv("C:\\Users\\KUNAL\\Downloads\\AWCustomers.csv") 
print(df.shape)
df.head()
```

```python
df.info()
df.describe(include='all')
```

```python

# Select relevant columns
selected_columns = [
    'BirthDate', 'Education', 'Occupation', 'Gender', 'MaritalStatus',
    'HomeOwnerFlag', 'NumberCarsOwned', 'NumberChildrenAtHome',
    'TotalChildren', 'YearlyIncome'
]

df_selected = df[selected_columns]
df_selected.head()
```

```python
from sklearn.preprocessing import LabelEncoder

# Convert BirthDate to Age
df_selected['Age'] = 2025 - pd.to_datetime(df_selected['BirthDate']).dt.year
df_selected.drop('BirthDate', axis=1, inplace=True)

# Encode categorical variables
le = LabelEncoder()
for col in ['Education', 'Occupation', 'Gender', 'MaritalStatus']:
    df_selected[col] = le.fit_transform(df_selected[col])

df_selected.head()
```

```python
import pandas as pd
from scipy.stats import f_oneway  

# Group incomes by number of cars owned
groups = [df_selected[df_selected['NumberCarsOwned'] == n]['YearlyIncome'] 
          for n in df_selected['NumberCarsOwned'].unique()]

# Perform ANOVA
f_stat, p_val = f_oneway(*groups)
print("F-statistic:", f_stat)
print("P-value:", p_val)

if p_val < 0.05:
    print(" Significant difference in average income across different car ownership groups.")
else:
    print(" No significant difference in income across groups.")
```

```python
import seaborn as sns
import matplotlib.pyplot as plt

sns.boxplot(x='NumberCarsOwned', y='YearlyIncome', data=df_selected)
plt.title('Income Distribution by Number of Cars Owned')
plt.show()
```

```python
edu_car = df_selected.groupby('Education')['NumberCarsOwned'].value_counts(normalize=True).unstack().fillna(0) * 100
print(edu_car)
```

```python
edu_car.plot(kind='bar', stacked=True, figsize=(8,5))
plt.title("Car Ownership % by Education Level")
plt.xlabel("Education")
plt.ylabel("Percentage")
plt.legend(title="Number of Cars Owned")
plt.show()
```

```python
sns.histplot(data=df_selected, x='Age', hue='NumberCarsOwned', multiple='stack', bins=20)
plt.title("Age Distribution by Number of Cars Owned")
plt.xlabel("Age")
plt.ylabel("Count")
plt.show()
```

```python
marital_car = df_selected.groupby('MaritalStatus')['NumberCarsOwned'].mean()
print(marital_car)

sns.barplot(x='MaritalStatus', y='NumberCarsOwned', data=df_selected, estimator='mean')
plt.title("Average Number of Cars Owned by Marital Status")
plt.ylabel("Average Cars Owned")
plt.show()
```
