```python
employees = {
    "Employee_ID": [101, 102, 103, 104, 105],
    "Name": ["Alice", "Bob", "Charlie", "Diana", "Edward"],
    "Department": ["HR", "IT", "IT", "Marketing", "Sales"],
    "Age": [29, 34, 41, 28, 38],
    "Salary": [50000, 70000, 65000, 55000, 60000],
    "Years_of_Experience": [4, 8, 10, 3, 12],
    "Joining_Date": ["2020-03-15", "2017-07-19", "2013-06-01", "2021-02-10", "2010-11-25"],
    "Gender": ["Female", "Male", "Male", "Female", "Male"],
    "Bonus": [5000, 7000, 6000, 4500, 5000],
    "Rating": [4.5, 4.0, 3.8, 4.7, 3.5]
}

emp_df = pd.DataFrame(employees)
```

```python
print("\nShape of DataFrame (rows, columns):", emp_df.shape)
print("\nDataFrame Info:")
print(emp_df.info())
print("\nDescriptive Statistics:\n", emp_df.describe())
print("\nFirst 5 rows:\n", emp_df.head())
print("\nLast 3 rows:\n", emp_df.tail(3))
```

```python
avg_salary = emp_df["Salary"].mean()
total_bonus = emp_df["Bonus"].sum()
youngest_age = emp_df["Age"].min()
highest_rating = emp_df["Rating"].max()
```

```python
print("\nAverage Salary:", avg_salary)
print("Total Bonus:", total_bonus)
print("Youngest Age:", youngest_age)
print("Highest Rating:", highest_rating)
```

```python
sorted_df = emp_df.sort_values(by="Salary", ascending=False)
print("\nSorted by Salary (Descending):\n", sorted_df)
```

```python
def categorize(rating):
    if rating >= 4.5:
        return "Excellent"
    elif 4.0 <= rating < 4.5:
        return "Good"
    else:
        return "Average"
```

```python
emp_df["Performance_Category"] = emp_df["Rating"].apply(categorize)
print("\nAfter adding Performance_Category:\n", emp_df)
```

```python
print("\nMissing values in DataFrame:\n", emp_df.isnull().sum())
emp_df.rename(columns={"Employee_ID": "ID"}, inplace=True)
print("\nAfter renaming column:\n", emp_df.head())
```

```python
print("\nEmployees with more than 5 years of experience:\n", emp_df[emp_df["Years_of_Experience"] > 5])
```

```python
print("\nEmployees in IT department:\n", emp_df[emp_df["Department"] == "IT"])
```

```python
emp_df["Tax"] = emp_df["Salary"] * 0.10
print("\nAfter adding Tax column:\n", emp_df)
```

```python
emp_df.to_csv("modified_employees.csv", index=False)
print("\nModified DataFrame saved as 'modified_employees.csv'")
```
